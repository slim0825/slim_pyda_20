### Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
from sympy import diff, symbols, cos, sin
from scipy.misc import derivative
```

```python
x = symbols('x')
f = x**5 + 4 * sin(2 * x) + cos(3 * x + 3)
def ff(x):
    return (x**5 + 4 * sin(2 * x) + cos(3 * x + 3))


df = diff(f)
a = derivative(ff, 1, n = 1)
print(f"первая производная в точке х=1 равна {round(a, 3)}")
df
```

```python
ddf = diff(df)
b = derivative(ff, 1, n = 2)
print(f"вторая производная в точке х=1 равна {round(b, 3)}")
ddf
```

### Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.utilities.lambdify import lambdify
```

```python
f2 = (sin(2 * x + 1))** 5
df2 = diff(f2)
ddf2 = diff(df2)

xs = np.linspace(-5, 5, 500)
y = lambdify(x, df2)          ### график первой производной
z = lambdify(x, ddf2)         ### график второй производной
plt.figure(figsize=(10, 10))

plt.plot(xs, [y(x) for x in xs])
plt.plot(xs, [z(x) for x in xs])
```
