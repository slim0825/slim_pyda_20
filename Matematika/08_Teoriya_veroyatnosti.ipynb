```python
import numpy as np
import scipy.stats as sts
from scipy.stats import norm
import matplotlib.pyplot as plt
```

### Уровень 0:   
   
Задание 1:   
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
mu_arr = np.linspace(1.0, 6.0, 5)
var_arr = np.linspace(0.1, 0.6, 5)
x = np.linspace(0, 8, 100)
def f(x, norm):    
    pdf = norm.pdf(x)
    return pdf
```

```python
fig, axes = plt.subplots(figsize = (15, 6))
plt.title('Плотность нормального распределения вероятности')
axes.set_xlabel('x', fontsize = 12)
axes.set_ylabel('$f(x)$', fontsize = 12)
for mu, var in zip(mu_arr, var_arr):
    norm = sts.norm(loc = mu, scale = var)
    axes.plot(x, f(x, norm), label = [mu, var])
    axes.legend(loc='best', frameon=False)
plt.show
```

```python
lam_arr = np.linspace(0.1, 6.0, 5)
x = np.linspace(0, 10, 100)
def f(x, expon):    
    pdf = expon.pdf(x)
    return pdf
```

```python
fig, axes = plt.subplots(figsize = (15, 6))
plt.title('Плотность экспоненциального распределения вероятности')
axes.set_xlabel('x', fontsize = 12)
axes.set_ylabel('$f(x)$', fontsize = 12)

for lam in lam_arr:
    expon = sts.expon(lam)
    axes.plot(x, f(x, expon), label = lam)
    axes.legend(loc='best', frameon=False)
plt.show
```

```python
n_arr = np.linspace(1.0, 10.0, 5)
x = np.linspace(-8, 8, 100)
def f(x, t):    
    pdf = t.pdf(x)
    return pdf
```

```python
fig, axes = plt.subplots(figsize = (15, 6))
plt.title('Плотность распределения вероятности по Стьюденту ')
axes.set_xlabel('x', fontsize = 12)
axes.set_ylabel('$f(x)$', fontsize = 12)
for n in n_arr:
    t = sts.t(n)
    axes.plot(x, f(x, t), label = n)
    axes.legend(loc='best', frameon=False)
plt.show
```
