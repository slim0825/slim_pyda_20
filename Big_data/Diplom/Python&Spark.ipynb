# Практика Python

```python
import pandas as pd
```

### 1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
ratings = pd.read_csv('u.data.csv', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
ratings.head()
```

```python
movies = pd.read_csv('u.item.csv', sep='|', encoding='latin-1', names = ['movie_id', 'movie_title', 'release_date', \
                                                                               'video_release_date', 'IMDb_URL', 'unknown',\
                                                                               'Action', 'Adventure', 'Animation', \
                                                                               'Childrens', 'Comedy', 'Crime', \
                                                                               'Documentary', 'Drama', 'Fantasy', \
                                                                               'Film-Noir', 'Horror', 'Musical', 'Mystery',\
                                                                               'Romance', 'Sci-Fi', 'Thriller', 'War', \
                                                                               'Western'])
movies.head()
```

### 2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
ratings['user_id'].value_counts(ascending=False)
```

```python
rating = ratings.groupby('user_id').count().reset_index()
rating_max = rating.sort_values('item_id', ascending = False).head(1)
rating_max
```

### 3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
df = []
df = ratings[ratings['user_id'] == 405].join(movies.set_index('movie_id'), how='inner', on='item_id')
df.head()
```

### 4. Добавьте к датафрейму из задания 3 столбцы:

#### 4.1 По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет


```python
df['relese_year'] = df.apply(lambda row: int(row['release_date'][-4:]), axis=1)
df
```

#### 4.2 Столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

```python
rat_cnt = ratings.groupby('item_id').agg('count').loc[:,['rating']]
rat_cnt.rename({'rating': 'rating_cnt'}, axis=1, inplace=True)
rat_sum = ratings.groupby('item_id').agg('sum').loc[:,['rating']]
rat_sum.rename({'rating': 'rating_sum'}, axis=1, inplace=True)
rating_1 = rat_cnt.join(rat_sum, how='inner')
rating_1.reset_index()
```

```python
df = df.join(rating_1, how = 'inner', on='item_id')
df.head()
```

### 5. Сформируйте X_train, X_test, y_train, y_test
### 6. Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
```

```python
# Сформировать X и y из df, выбрав нужные колонки
X, y = df[['Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', \
        'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', \
        'relese_year', 'rating_cnt', 'rating_sum']], df[['rating']]
```

```python
# Разбить данные на данные для обучения и проверки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
model = LinearRegression() ### Создать и обучить модель LinearRegression
```

```python
model.fit(X_train, y_train)
print(model.coef_, model.intercept_)
```

### 7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test) ### Коэффициент детерменизма
```

```python
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

## Вторая часть практики Python связана со Spark’ом

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
```

### 8. Загрузить данные в spark

```python
movies = spark.read.csv('u.item_spark.csv', header=None, inferSchema=True, sep='|')
movies.show(5)
```

```python
movies = movies.toDF('movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown',\
                     'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama',\
                     'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery','Romance', 'Sci-Fi', 'Thriller',\
                     'War', 'Western')
movies.show(5)
```

```python
ratings = spark.read.csv('u.data_spark.csv', header=None, inferSchema=True, sep='\t')
ratings.show(5)
```

```python
ratings = ratings.toDF('user_id', 'item_id', 'rating', 'timestamp')
ratings.show(5)
```

### 9. Средствами спарка вывести среднюю оценку для каждого фильма

```python
avg_rat = ratings.groupBy('item_id')
avg_rat = avg_rat.avg('rating').select('item_id', 'avg(rating)')
avg_rat.join(movies, on=avg_rat['item_id'] == movies['movie_id'], how = 'inner').select('movie_title', 'avg(rating)').show()
```

### 10. Посчитайте средствами спарка среднюю оценку для каждого жанра

```python
genre = ratings.join(movies, on=ratings['item_id'] == movies['movie_id'], how = 'inner')
genre.where(genre['Action'] == 1).groupBy('Action').avg('rating').show()
genre.where(genre['Adventure'] == 1).groupBy('Adventure').avg('rating').show()
genre.where(genre['Animation'] == 1).groupBy('Animation').avg('rating').show()
genre.where(genre['Childrens'] == 1).groupBy('Childrens').avg('rating').show()
genre.where(genre['Comedy'] == 1).groupBy('Comedy').avg('rating').show()
genre.where(genre['Crime'] == 1).groupBy('Crime').avg('rating').show()
genre.where(genre['Documentary'] == 1).groupBy('Documentary').avg('rating').show()
genre.where(genre['Drama'] == 1).groupBy('Drama').avg('rating').show()
genre.where(genre['Fantasy'] == 1).groupBy('Fantasy').avg('rating').show()
genre.where(genre['Film-Noir'] == 1).groupBy('Film-Noir').avg('rating').show()
genre.where(genre['Horror'] == 1).groupBy('Horror').avg('rating').show()
genre.where(genre['Musical'] == 1).groupBy('Musical').avg('rating').show()
genre.where(genre['Mystery'] == 1).groupBy('Mystery').avg('rating').show()
genre.where(genre['Romance'] == 1).groupBy('Romance').avg('rating').show()
genre.where(genre['Sci-Fi'] == 1).groupBy('Sci-Fi').avg('rating').show()
genre.where(genre['Thriller'] == 1).groupBy('Thriller').avg('rating').show()
genre.where(genre['War'] == 1).groupBy('War').avg('rating').show()
genre.where(genre['Western'] == 1).groupBy('Western').avg('rating').show()
```
