### 4. Построить график количества фильмов по годам

```python
from datetime import datetime
```

```python
datas = data['release_date']
datas = datas.dropna()
```

```python
data_data = []
for x in datas:
    date = datetime.strptime( x, '%d-%b-%Y' )
    result = date.strftime('%Y')
    data_data.append(int(result))
```

```python
df = pd.DataFrame({'col0':data_data, 'col1':data_data})
df = df.groupby('col0').count().reset_index()
df.columns = ['date','count']
df
```

```python
fig, axs = plt.subplots(figsize=(15, 10))
sns.barplot(x=df['count'], y=df['date'], palette='hls')

plt.xlabel('Количество фильмов')
plt.ylabel('Год')
plt.title("Распределение фильмов по годам")
plt.grid()
plt.ylim(1920,2000)
plt.show()
```

### 5. Построить график количества оценок по месяцам и годам (преобразование timestamp в дату смотерть [тут](https://stackoverflow.com/questions/45227380/convert-unix-epoch-time-to-date-in-google-sheets))

```python
data = pd.read_csv('u.data.csv', sep='\t', header=None)
data.head()
```

```python
data.columns = ['user_id', 'item_id', 'rating', 'timestamp']
data.head()
```

```python
data.info()
```

```python
def column_of_date(row):                              ### Даты оценок фильмов
    return datetime.fromtimestamp(row['timestamp'])
data['date'] = data.apply(column_of_date, axis=1)

def column_of_year_and_month(row):                    ### Год и месяц фильмов
    return row['date'].year, row['date'].month
data['ym'] = data.apply(column_of_year_and_month, axis=1)

data.head()
```

```python
data_5 = data[['rating','ym']]
result = data_5.groupby('ym').count().reset_index()
result.columns = ['date','count']
result
```

```python
fig, axs = plt.subplots(figsize=(15, 10))
sns.barplot(x=result['count'], y=result['date'], palette='hls')

plt.xlabel('Количество оценок')
plt.ylabel('Год и месяц')
plt.title("Распределение оценок по годам и месяцам")
plt.grid()
plt.show()
```

### 6. Выявить top-5 самых активных пользователей (больше всего оценок) за последние 3 месяца

```python
data_5 = data[['rating','ym']]
result = data_5.groupby('ym').count().reset_index()
result.columns = ['date','count']
tail = result.tail(3)
date_6 = tail[['date']]
date_6
```

```python
user_data = data[(data['ym'] == date_6['date'][5]) | (data['ym'] == date_6['date'][6]) | (data['ym'] == date_6['date'][7])]
user_datas = user_data.groupby('user_id').count().reset_index()
user_ratings = user_datas.sort_values(by='rating', ascending=False).head()
top_users = user_ratings[['user_id','rating']]
top_users.columns = ['user_id','count_rating']
top_users
```
