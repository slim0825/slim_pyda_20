### Задание 1   

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:   
   
* если источник traffic_source равен yandex или google, то в source_type ставится organic   
* для источников paid и email из России - ставим ad   
* для источников paid и email не из России - ставим other   
* все остальные варианты берем из traffic_source без изменений   

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
# если источник traffic_source равен yandex или google, то в source_type ставится organic
log.loc[log.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'
log.head()
```

```python
# для источников paid и email из России - ставим ad
log.loc[(log.traffic_source.isin(['paid', 'email'])) & (log.region == 'Russia'), 'source_type'] = 'ad'
log.head()
```

```python
# для источников paid и email не из России - ставим other
log.loc[(log.traffic_source.isin(['paid', 'email'])) & (log.region != 'Russia'), 'source_type'] = 'other'
log.head()
```

```python
# все остальные варианты берем из traffic_source без изменений
log.loc[(~log.traffic_source.isin(['yandex', 'google'])) & (~log.traffic_source.isin(['paid', 'email'])) , 'source_type']
log.head()
```

### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
df = pd.read_csv('ml-latest-small/ratings.csv')
df.head()
```

```python
visits = df.groupby('userId').agg({'timestamp':['min', 'max', 'count']}).reset_index()
visits.iloc[:5]
```

```python
visits_filtred = visits[visits['timestamp']['count'] > 100]
visits_filtred.iloc[:5]
```

```python
# среднее время жизни каждого пользователя
visits_filtred['life_time'] = visits_filtred[('timestamp', 'max')] - visits_filtred[('timestamp', 'min')]
visits_filtred.iloc[:5]
### не понимаю что за ошибка снизу, хотя считает всё верно
```

```python
avg_life = visits_filtred['life_time'].mean()
print(f'Среднее время жизни пользователей, которые выставили более 100 оценок = {int(avg_life)}')
```

```python
### timestamp Unixtime - Количество секунд, прошедших с 1 января 1970 года по UTC
print(f'Среднее время жизни пользователей, которые выставили более 100 оценок = {40080507 / 60 / 60 / 24} дня')
```
