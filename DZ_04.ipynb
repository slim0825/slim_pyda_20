```python
documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
            ]
 
directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
               }
### 1. Пользователь по команде "p" может узнать владельца документа по его номеру

def people(numbers):
    for doc_number in documents:
        if doc_number["number"] == numbers:
            print('Владелец документа:', doc_number["name"])
            break
    else:
        print('Документ не найден в базе')

### 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
        
def shelf(numbers):
    break_marker = False
    for shelf_directories in directories.items():
        for doc_numbers in shelf_directories[1]:
            if doc_numbers == numbers:
                print('Документ хранится на полке:', shelf_directories[0])
                break_marker = True
                break
                if break_marker == True:
                    break
                else:
                    print('Документ не найден в базе')

### 3. Пользователь по команде "l" может увидеть полную информацию по всем документам                    
                    
def people_list():
    for shelf in directories:
        for num in directories[shelf]:
            found = next(e for e in documents if e['number']==num)
            print(f'№: {num}, тип: {found["type"]}, владелец: {found["name"]}, полка хранения: {shelf}')
                      
### 4. Пользователь по команде "ads" может добавить новую полку
                
def add_shelf():
    shelf_number = input("Введите номер полки: ")
    for shelf in directories.items():
        if shelf_number in shelf[0]:
            print('Такая полка уже существует. Текущий перечень полок: ', list(directories.keys()))
            break
    else:
        directories[shelf_number] = []
        print('Полка добавлена. Текущий перечень полок:', list(directories.keys()))

### 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)        
        
def del_shelf():
    shelf_number = input("Введите номер полки:")
    if shelf_number not in directories.keys():
        print('Такой полки не существует. Текущий перечень полок:', list(directories.keys()))
    elif len (directories[shelf_number]) > 0:
        print ('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', list(directories.keys()))
    else:
        del(directories[shelf_number])
        print('Полка удалена. Текущий перечень полок:', list(directories.keys()))
                    
while True:
    command = input('Введите команду: ')
    if command == 'p':
        people(input('\nВведите номер документа:'))
    elif command == 's':
        shelf(input('\nВведите номер документа:'))
    elif command == 'l':
        people_list()
    elif command == 'ads':
        add_shelf()
    elif command == 'ds':
        del_shelf()
    elif command == 'q':
        break
    else:
          print('Вы ввели некорректную команду, повторите ввод.')
```
